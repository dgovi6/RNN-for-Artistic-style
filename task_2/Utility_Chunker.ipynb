## Chunker mask_var

Le immagini sono divise in chunks quadrati il cui lato è pari al lato minore dell'immagine originale. 
I chunks sono distribuiti uniformemente e sono in numero tale da coprire l'immagine con un massimo del 50% di sovrapposizione su ogni dimensione.

```python
# funzione per calcolare la distribuzione di chunks di lato mask_size lungo un lato lungo dim_len
def find_positions(dim_len, mask_size = 224):

    x = dim_len
    a = mask_size
    pos = []
    
    # calcolare numero di chunks che soddisfano vincolo di sovrapposizione
    nsteps = int(math.floor( dim_len /(mask_size/2.)))
    nclust = nsteps - 1
    
    # controllo di sicurezza
    if nclust < 1: raise IndexError
    
    for c in range(nclust):
        # calcolare la posizione per il chunk c-esimo
        p = dim_len/nsteps + c*dim_len/nsteps - mask_size/2
        pos.append(p)
    return pos


# funzione per ottenere i chunks da immagine
def chunkify_var(image):

    # dimensioni immagine
    shape = list(image.shape)
    
    # lista output
    chunk_list = []

    # lato minore come lato chunk
    dim = np.argmin(shape[:2])
    mask_size = shape[dim]
    
    # calcolo posizioni chunks
    vpos = find_positions(shape[0], mask_size)
    hpos = find_positions(shape[1], mask_size)
    
    # estrazione chunk per ogni posizione griglia
    for x in vpos:
        for y in hpos:
            x = int(x)
            y = int(y)
            mask_size= int(mask_size)
            chunk = image[x:x+mask_size,y:y+mask_size,:]
            chunk_list.append(chunk)
    return chunk_list
```

## Chunker Mask5
Le immagini sono divise in chunks usando una griglia 2x2+1 di chunks aventi come lato metà del lato minore e centrata sull'immagine.

```python
# funzione per calcolare la distribuzione di chunks di lato mask_size lungo un lato lungo dim_len secondo la griglia 2z2+1
def find_positions_mask5(dim_len, mask_size = 224):
    pos = []
    
    # centro del lato
    c = dim_len / 2
    
    # margini
    p1 = c - mask_size
    p2 = c - mask_size/2
    p3 = c
    
    pos = [p1,p2,p3]
    return pos


def chunkify_mask5(image):
    
    # dimensioni immagine
    shape = list(image.shape)
    
    # lista output
    chunk_list = []
    
    # mezzo lato minore come dimensione chunk
    dim = np.argmin(shape[:2])
    mask_size = shape[dim]/2
    
    # calcolo posizioni chunks
    vpos = find_positions_mask5(shape[0], mask_size)
    hpos = find_positions_mask5(shape[1], mask_size)
    
    # estrazione chunks gliglia 2x2
    for v in [vpos[0],vpos[2]]:
        for h in [hpos[0],hpos[2]]:
            v = int(v)
            h = int(h)
            mask_size= int(mask_size)
            chunk = image[v:v+mask_size,h:h+mask_size,:]
            chunk_list.append(chunk)
            
    # posizionamento chunk centrale
    vc = int(vpos[1])
    hc = int(hpos[1])
    chunk_list.append(image[vc:vc+mask_size,hc:hc+mask_size,:])
    
    return chunk_list
```
